```python
def cluster(feature):
    
    _, axes = plt.subplots(figsize=(20,8))
    sns.kdeplot(x=df[feature], y=df["TotalMntSpent"], hue =df["Cluster"], kind="kde",height=10, palette=color)
    axes.set_title(f"\nClustering Profile On {feature}\n",fontsize=25)
    axes.set_ylabel("Spending",fontsize=20)
    axes.set_xlabel(f"\n{feature}",fontsize=20)
    
    sns.despine(left=True, bottom=True)
    plt.show()
```

```python
col = ["Kidhome" , "Age" , "Education_Level" , "Partner" , "Customer Loyalty"]

for i in col:
    print(f"Characteristics of clusters based on spending and {i} :")
    cluster(i)
```
